# Creating a Filter, Edge Detection

### Import resources and display image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline

# Read in the image
images = [mpimg.imread('images/curved_lane.jpg'), mpimg.imread('images/bridge_trees_example.jpg')]

f, (ax, bx) = plt.subplots(1, 2, figsize=(20, 10))
ax.imshow(images[0])
ax.set_title('curved_lane.jpg')
bx.imshow(images[1])
bx.set_title('bridge_trees_example.jpg')
```

### Convert the image to grayscale

```python
def apply_filter(image):    
    # Convert to grayscale for filtering
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # plt.imshow(gray, cmap='gray')
    
    sobel_y = np.array([[ -1, -2, -1],
                       [ 0, 0, 0],
                       [ 1, 2, 1]])
    
    # Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
    filteredy_image = cv2.filter2D(gray, -1, sobel_y)

    sobel_x = np.array([[ -1, 0, 1], 
                       [ -2, 0, 2], 
                       [ -1, 0, 1]])
    filteredx_image = cv2.filter2D(gray, -1, sobel_x)
    

    f, (ax, bx) = plt.subplots(1, 2, figsize=(20, 10))
    ax.imshow(filteredy_image, cmap='gray')
    ax.set_title('sobel_y')
    bx.imshow(filteredx_image, cmap='gray')
    bx.set_title('sobel_x')
    
```
